## Exercise 1 - Simple math 🔢

1. Create the variables `a` and `b` and assign to them the values `252` and `151`, respectively.
2. Use the basic math operators (`+`, `-`, `*`, `/`) to calculate the sum, difference, product and quotient of `a` and `b`. Save each result in a separate variable.
3. Is the result of the division correct? If not, think about which data type is required.
4. Use `System.out.println` to output the results.

## Exercise 2 - Friendly strings 🙂

1. Create a variable `greeting` and store a pleasant greeting of your choice, e.g. "Hello!".
2. Output the greeting.
1. Create a second variable `name` and store your own name, e.g. "Philipp".
2. Using the variables `greeting` and `name`, output a sentence containing the greeting and your name, e.g. "Hello! My name is Philipp!".

## Exercise 5 - FizzBuzz



FizzBuzz is a popular coding exercise that started as a way to quickly assess a person's programming skills in a job interview. The idea is simple, but there are multiple ways to solve the task. We want to go over the numbers 1 through 20 and output something based on the following rules:

- If the number is divisible by 3, output "Fizz".
- If the number is divisible by 5, output "Buzz".
- If the number is divisible by both 3 and 5, output "FizzBuzz".
- If the number is neither divisible by 3 nor by 5, output the number itself.
